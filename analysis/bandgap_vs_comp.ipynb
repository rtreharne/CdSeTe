```python
import os
import sys
import csv
CdTe = os.path.abspath(os.path.join('CdTe_prof.txt'))
CdSe = os.path.abspath(os.path.join('CdSe_prof.txt'))
Eg = os.path.abspath(os.path.join('160628-B_bandgap.txt'))


```

```python
def get_data(path):
    f = open (path, 'rb')
    reader = csv.reader(f)
   
    data = {}
    data["header"] = []
       
    for line in reader:
        item = line[0].split("\t")
        try:
           
            array = [float(x) for x in item]
            label = '{0},{1}'.format(array[0], array[1])
            data[label] = {'x': array[0], 'y': array[1], 'z':array[2]}
        except:
            data["header"].append(item)
    
    return data
```

```python
Egdat = get_data(Eg)
CdTedat = get_data(CdTe)
CdSedat = get_data(CdSe)
```

```python
def calc_comp(CdTe, CdSe, Eg):
    print len(CdTe), len(CdSe)
    new = {}
    for key in CdTe:
        try:
            total = CdSe[key].get('z')+CdTe[key].get('z')
            new[key] = {'y': CdTe[key].get('y', None),
                        'x': CdTe[key].get('x', None),
                        'comp': (CdSe[key].get('z')/total)*100,
                        'Eg': Eg[key].get('z', None),
                        'total': total}
        except:
            continue
    
    return new
    
```

```python
new = calc_comp(CdTedat, CdSedat, Egdat)
```

```python
new
```

```python
x = []
y = []
z = []
for key in new:
    if (new[key].get('total', None) > 400):
        x.append(new[key].get('comp', None))
        y.append(new[key].get('Eg', None))
        z.append(new[key].get('total', None))
    
```

```python
fig = figure(figsize(11,7))
ax = fig.add_subplot(111)
L1 = ax.scatter(x, y, c=z, s=100)
ax.tick_params(labelsize=14)
CB = colorbar(L1)
CB.set_label('Film Thickness (nm)', fontsize=20)
CB.ax.tick_params(labelsize=12)
#ax.set_ylim([1.5,1.9])
ax.set_xlabel('% Se', fontsize=20)
ax.set_ylabel('Band Gap (eV)', fontsize=20)
```
